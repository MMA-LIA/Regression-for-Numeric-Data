```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from sklearn.model_selection import train_test_split
from tensorflow.keras import optimizers
from tensorflow.keras.layers import Dense, Activation,Input
import operator
```

# Linear Regression

```python
# Simulated Data Generation 
x_data= np.linspace(-0.5, 0.5, 200)[:, np.newaxis]
noise= np.random.normal(0, 0.02, x_data.shape)
y_data=np.square(x_data) + noise

# Linear Regression model fitting
lg=LinearRegression()
lg.fit(x_data, y_data)

# Make Prediction  on x_data
y_pred=lg.predict(x_data)


#Mean Squared Error
print("Mean Squared Error: ",  mean_squared_error(y_data, y_pred))

#Plot Prediction cureve
plt.scatter(x_data, y_data)
plt.plot(x_data, y_pred, color='red')
plt.title("Predictoin Curve with Linear Regression", fontsize=16)
plt.xlabel("X Data", fontsize=14)
plt.ylabel("Y Data", fontsize=14)
plt.show()
```

#Polynomial Model

```python
# Simulated Data Generation 
x_data= np.linspace(-0.5, 0.5, 200)[:, np.newaxis]
noise= np.random.normal(0, 0.02, x_data.shape)
y_data=np.square(x_data) + noise

# Apply polynimial features of degree 2 on x_data
polynomial_feature= PolynomialFeatures(degree=2)
x_poly= polynomial_feature.fit_transform(x_data)

# Model fitting with polynomial of degree 2
polyreg= LinearRegression()
polyreg.fit(x_poly, y_data)

# Make prediction on x_data
y_pred=polyreg.predict(x_poly)

# Mean squared error
print("mean quared error: ", mean_squared_error(y_data, y_pred))

# Plot Prediction cureve
plt.scatter(x_data, y_data)
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x_data,y_pred), key=sort_axis)
x_data, y_pred = zip(*sorted_zip)
plt.plot(x_data, y_pred, color='red')
plt.title("Predictoin Curve for Polynomial Model of Degree 2", fontsize=16)
plt.xlabel("X Data", fontsize=14)
plt.ylabel("Y Data", fontsize=14)
plt.show()
```

# Layers Model DNN

```python
# Simulated Data Generation 
x_data= np.linspace(-0.5, 0.5, 200)[:, np.newaxis]
noise= np.random.normal(0, 0.02, x_data.shape)
y_data=np.square(x_data) + noise

# split data into 80% training and 20% testing
X_train,X_test,y_train,y_test=train_test_split(x_data,y_data, test_size=0.2, random_state=0)

# Model Architecture
model = Sequential()
model.add(Input(1))
model.add(Dense(6,activation='relu'))
model.add(Dense(1))
model.compile(optimizer='Adam',loss='mean_squared_error')
print(model.summary())

# fit model for 50 epoch and batch size 8
history=model.fit(X_train,y_train,batch_size=8,epochs=50)

# Make prediction
y_pred = model.predict(X_test)

# Mean squared Error
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))

# plot Training Loss
plt.plot(history.history['loss'],marker='o',linewidth=3, color="red")
plt.title("Training Loss", fontsize=16)
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("Mean Square Error (MSE)", fontsize=14)
plt.show()
```
